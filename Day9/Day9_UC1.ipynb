```python
# Pull the complete Webpage into a TREE 
from lxml import html
import requests

page1 = requests.get('https://www.alexa.com/topsites/countries/AE')
tree1 = html.fromstring(page1.content)
page1.content
```

```python
tree1
```

```python
# Pull the Table using xpath into a file mee.txt
snip1 = tree1.xpath('//*[@id="alx-content"]/div/section/div[1]/section[2]/span/span/div/div/div[2]')
import sys

for tag1 in snip1:
    f = open('mee.txt', 'w')
    print(tag1.text_content(), file=f)
    f.close()
```

```python
# Remove all BLANK lines and Spaces and write to mee1.txt
with open("mee.txt" ,"r") as data:
    a2 = data.read()
    a2 = a2.split("\n")
    a3 = open("mee" + "1.txt","w")
    outp = [] # Creation of 
    for line in a2:
        line=line.strip()
        if line:
#            line = line.replace("  ",",")
#            line = line.replace('\t', ",")
            outp.append(line)
    a3.write('\n'.join(outp))
    a3.close()
```

```python
# Add SiteRank word to the Line 1 of the file and write to mee2.txt
open("mee2.txt", "w").write("SiteRank\n" + open("mee1.txt").read())
```

```python
# Remove unwanted lines 4, 6, 8 and 10 and write to mee3.txt
infile = open('mee2.txt','r').readlines()
with open('mee3.txt','w') as outfile:
    for index,line in enumerate(infile):
        if index != 3:
            if index != 5:
                if index != 7:
                    if index != 9:
                        outfile.write(line)
```

```python
# Add ; to end of all the lines and write to mee4.txt (will be used later for CSV conversion)
with open('mee3.txt', 'r') as istr:
    with open('mee4.txt', 'w') as ostr:
        for line in istr:
            line = line.rstrip('\n') + ';'
            print(line, file=ostr)
```

```python
z = open("mee4.txt").read().split("\n")
#z = z.split("\n")

sixlines = range(0,len(z),6)
with open('mee5.txt', 'w') as ostr:
    for num, line in enumerate(z):
        if num in sixlines:
            print (' '.join(z[num:num+6]), file=ostr)
```
